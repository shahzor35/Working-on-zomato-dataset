Extracting Indian restaurants from zomato data

```python
indian_res = result_after_drop[result_after_drop['Currency'] == 'Indian Rupees(Rs.)']
pd.DataFrame(indian_res)
```

```python
# We want to know few columns of the Indian restaurant that is having online delivery
indian_res[indian_res['Has Online delivery'] == 'Yes'].count()
```

```python
# We want to know the count of the restaurant name which is having online delivery
indian_res[indian_res['Has Online delivery'] == 'Yes']['Restaurant Name'].count()
```

```python
# We want to know the percentage of the restaurants having online delivery
has_online = np.sum(indian_res['Has Online delivery'] == 'Yes')
print("Total how many restaurants has online delivery",has_online)
percentage_has_online = ((has_online)/len(indian_res)) * 100
print("The percentage of restaurants having online delivery is : ",percentage_has_online)
```

```python
# We want to know how many Indian cities are there in Indian restaurant data set
pd.DataFrame(indian_res.groupby('City'))
```

```python
# We want to know the average cost of two in Indian cities
pd.DataFrame(indian_res.groupby('City')['Average Cost for two'])
```

```python
# We want to know the average cost of two in Indian cities maximum value
pd.DataFrame(indian_res.groupby('City')['Average Cost for two'].max())
```

```python
# We want to know the average cost of two in Indian cities maximum , minumum values
pd.DataFrame(indian_res.groupby('City')['Average Cost for two'].agg([max, min]))
```
